## 5. Data Analysis & Visualisasi

### 5.1. Analisis untuk Pertanyaan 1: Pengaruh `EX_level_demand` terhadap `Hourly_Rate`

```python
# Standarisasi nama kolom di awal sel untuk memastikan konsistensi
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')

# --- Data Cleaning untuk kolom yang relevan ---
# 1. Cleaning 'Hourly_Rate'
# Mengisi missing values di 'Hourly_Rate' dengan modus sebelum membersihkan, atau setelah membersihkan jika memungkinkan.
# Karena akan dikonversi ke numerik, mengisi dengan modus string dulu lalu konversi.
# Atau lebih baik, kita coba konversi, lalu tangani NaN yang dihasilkan.

# Memastikan kolom 'hourly_rate' ada
if 'hourly_rate' in df.columns:
    # Menghapus karakter non-numerik seperti '$', ',', dan 'hr' (jika ada) dan strip spasi
    df['hourly_rate_cleaned'] = df['hourly_rate'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.replace('/hr', '', regex=False).str.strip()

    # Mengubah ke tipe data numerik. Gunakan errors='coerce' untuk mengubah nilai yang tidak dapat dikonversi menjadi NaN.
    df['hourly_rate_cleaned'] = pd.to_numeric(df['hourly_rate_cleaned'], errors='coerce')

    # Mengisi nilai NaN di 'hourly_rate_cleaned' dengan median
    # Pastikan median dihitung dari kolom yang sudah bersih dan numerik
    median_hourly_rate = df['hourly_rate_cleaned'].median()
    df['hourly_rate_cleaned'] = df['hourly_rate_cleaned'].fillna(median_hourly_rate)

    print(f"'Hourly_Rate' setelah cleaning dan imputasi median ({median_hourly_rate}):")
    print(df[['hourly_rate', 'hourly_rate_cleaned']].head())
    print(f"Jumlah nilai null di 'hourly_rate_cleaned' setelah imputasi: {df['hourly_rate_cleaned'].isnull().sum()}")
else:
    print("Kolom 'hourly_rate' tidak ditemukan. Pastikan nama kolom sudah benar.")


# 2. Cleaning 'EX_level_demand'
# Memastikan kolom 'ex_level_demand' ada
if 'ex_level_demand' in df.columns:
    # Mengisi missing values di 'ex_level_demand' dengan modus
    mode_ex_level = df['ex_level_demand'].mode()[0]
    df['ex_level_demand'] = df['ex_level_demand'].fillna(mode_ex_level)
    print(f"\n'EX_level_demand' setelah imputasi modus ({mode_ex_level}):")
    print(df['ex_level_demand'].value_counts())
    print(f"Jumlah nilai null di 'ex_level_demand' setelah imputasi: {df['ex_level_demand'].isnull().sum()}")
else:
    print("Kolom 'ex_level_demand' tidak ditemukan. Pastikan nama kolom sudah benar.")


# --- Analisis untuk Pertanyaan 1: Pengaruh EX_level_demand terhadap Hourly_Rate ---
# Menghitung rata-rata Hourly_Rate berdasarkan EX_level_demand
average_rate_by_ex_level = df.groupby('ex_level_demand')['hourly_rate_cleaned'].mean().sort_values(ascending=False).reset_index()
print("\n### Rata-rata Hourly Rate berdasarkan EX_level_demand:")
print(average_rate_by_ex_level)

# Memvisualisasikan rata-rata Hourly_Rate berdasarkan EX_level_demand
plt.figure(figsize=(10, 6))
sns.barplot(data=average_rate_by_ex_level, x='ex_level_demand', y='hourly_rate_cleaned', hue='ex_level_demand', palette='coolwarm', legend=False)
plt.title('Average Hourly Rate by Experience Level Demand')
plt.xlabel('Experience Level Demand')
plt.ylabel('Average Hourly Rate (USD)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Memvisualisasikan variasi Hourly_Rate untuk setiap EX_level_demand (Boxplot)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='ex_level_demand', y='hourly_rate_cleaned', hue='ex_level_demand', palette='coolwarm', legend=False)
plt.title('Variasi Hourly Rate (USD) by Experience Level Demand')
plt.xlabel('Experience Level Demand')
plt.ylabel('Hourly Rate (USD)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 5.3. Analisis untuk Pertanyaan 3: Pengaruh `EX_level_demand` terhadap `target_hourly_rate_usd`

### 5.3. Analisis untuk Pertanyaan 3: Faktor-faktor yang Mempengaruhi `Rating` Freelancer

```python
# Standarisasi nama kolom (sudah dilakukan di sel sebelumnya, tapi diulang untuk konsistensi)
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')

# --- Data Cleaning untuk kolom yang relevan ---
# 1. Cleaning 'rating'
# Kita akan mengisi nilai NaN di 'rating' dengan median. Sebelum itu, pastikan kolomnya numerik.
if 'rating' in df.columns:
    # Pastikan 'rating' adalah numerik, jika ada nilai non-numerik akan menjadi NaN
    df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
    median_rating = df['rating'].median()
    df['rating'] = df['rating'].fillna(median_rating)
    print(f"\n'Rating' setelah imputasi median ({median_rating}):")
    print(f"Jumlah nilai null di 'rating' setelah imputasi: {df['rating'].isnull().sum()}")
else:
    print("Kolom 'rating' tidak ditemukan. Pastikan nama kolom sudah benar.")

# 2. Cleaning 'feedback_num'
# Kolom ini sudah numerik, hanya perlu menangani NaN.
if 'feedback_num' in df.columns:
    median_feedback_num = df['feedback_num'].median()
    df['feedback_num'] = df['feedback_num'].fillna(median_feedback_num)
    print(f"\n'Feedback_Num' setelah imputasi median ({median_feedback_num}):")
    print(f"Jumlah nilai null di 'feedback_num' setelah imputasi: {df['feedback_num'].isnull().sum()}")
else:
    print("Kolom 'feedback_num' tidak ditemukan. Pastikan nama kolom sudah benar.")

# Kolom 'spent($)' sudah dibersihkan di analisis sebelumnya
# Kolom 'ex_level_demand' sudah dibersihkan di analisis sebelumnya

# --- Analisis untuk Pertanyaan 3: Faktor-faktor yang Mempengaruhi Rating Freelancer ---
# Memilih kolom-kolom numerik yang relevan untuk analisis korelasi dengan 'rating'
# Gunakan 'hourly_rate_cleaned' dan 'applicants_num_cleaned' yang sudah dibuat sebelumnya
correlation_cols = ['rating', 'feedback_num', 'spent($)', 'hourly_rate_cleaned', 'applicants_num_cleaned']

# Filter hanya kolom yang ada di DataFrame
existing_correlation_cols = [col for col in correlation_cols if col in df.columns]

if len(existing_correlation_cols) > 1:
    correlation_matrix = df[existing_correlation_cols].corr()
    print("\n### Matriks Korelasi dengan Rating Freelancer:")
    print(correlation_matrix['rating'].sort_values(ascending=False))

    # Visualisasi Korelasi
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix[['rating']], annot=True, cmap='coolwarm', fmt='.2f', cbar=False)
    plt.title('Korelasi Faktor-faktor dengan Rating Freelancer')
    plt.show()
else:
    print("Tidak cukup kolom yang relevan untuk menghitung korelasi dengan 'rating'.")

# Analisis dengan 'ex_level_demand' (variabel kategorikal)
# Menggunakan boxplot untuk melihat distribusi rating berdasarkan level pengalaman
if 'ex_level_demand' in df.columns and 'rating' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='ex_level_demand', y='rating', hue='ex_level_demand', palette='viridis', legend=False)
    plt.title('Distribusi Rating Berdasarkan Level Pengalaman yang Diminta')
    plt.xlabel('Level Pengalaman yang Diminta')
    plt.ylabel('Rating')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("Tidak dapat memvisualisasikan rating berdasarkan 'ex_level_demand' karena kolom tidak ditemukan.")
```

```python
# Standarisasi nama kolom (sudah dilakukan di sel sebelumnya, tapi diulang untuk konsistensi)
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')

# --- Data Cleaning untuk kolom yang relevan ---
# 1. Cleaning 'applicants_num'
# Kolom ini perlu diubah ke numerik. Kita perlu menangani string seperti '5+' atau 'None'.

if 'applicants_num' in df.columns:
    # Mengganti '5+' dengan 5, dan mengubah yang lain menjadi numerik, 'None' atau non-numerik lainnya menjadi NaN
    df['applicants_num_cleaned'] = df['applicants_num'].astype(str).str.replace('+', '', regex=False).str.strip()
    df['applicants_num_cleaned'] = pd.to_numeric(df['applicants_num_cleaned'], errors='coerce')

    # Mengisi nilai NaN di 'applicants_num_cleaned' dengan median
    median_applicants_num = df['applicants_num_cleaned'].median()
    df['applicants_num_cleaned'] = df['applicants_num_cleaned'].fillna(median_applicants_num)

    print(f"'Applicants_Num' setelah cleaning dan imputasi median ({median_applicants_num}):")
    print(df[['applicants_num', 'applicants_num_cleaned']].head())
    print(f"Jumlah nilai null di 'applicants_num_cleaned' setelah imputasi: {df['applicants_num_cleaned'].isnull().sum()}")
else:
    print("Kolom 'applicants_num' tidak ditemukan. Pastikan nama kolom sudah benar.")

# 2. Cleaning 'spent($)'
# Kolom ini sudah float, perlu menangani NaN.
if 'spent($)' in df.columns:
    # Mengisi nilai NaN di 'spent($)' dengan median
    median_spent = df['spent($)'].median()
    df['spent($)'] = df['spent($)'].fillna(median_spent)
    print(f"\n'Spent($)' setelah imputasi median ({median_spent}):")
    print(f"Jumlah nilai null di 'spent($)' setelah imputasi: {df['spent($)'].isnull().sum()}")
else:
    print("Kolom 'spent($)' tidak ditemukan. Pastikan nama kolom sudah benar.")


# 3. Cleaning 'category_1'
# Mengisi missing values di 'category_1' dengan modus.
if 'category_1' in df.columns:
    mode_category_1 = df['category_1'].mode()[0]
    df['category_1'] = df['category_1'].fillna(mode_category_1)
    print(f"\n'Category_1' setelah imputasi modus ({mode_category_1}):")
    print(f"Jumlah nilai null di 'category_1' setelah imputasi: {df['category_1'].isnull().sum()}")
else:
    print("Kolom 'category_1' tidak ditemukan. Pastikan nama kolom sudah benar.")


# --- Analisis untuk Pertanyaan 2: Kategori Pekerjaan dengan Permintaan dan Potensi Pendapatan Tertinggi ---
# Menghitung rata-rata applicants_num dan spent($) berdasarkan category_1
category_analysis = df.groupby('category_1').agg(
    avg_applicants=('applicants_num_cleaned', 'mean'),
    avg_spent=('spent($)', 'mean')
).reset_index()

# Mengambil 10 kategori teratas berdasarkan rata-rata pelamar
top_demand_categories = category_analysis.sort_values(by='avg_applicants', ascending=False).head(10)
print("\n### 10 Kategori Pekerjaan Teratas Berdasarkan Permintaan (Rata-rata Pelamar):")
print(top_demand_categories)

# Mengambil 10 kategori teratas berdasarkan rata-rata Spent($)
top_earning_categories = category_analysis.sort_values(by='avg_spent', ascending=False).head(10)
print("\n### 10 Kategori Pekerjaan Teratas Berdasarkan Potensi Pendapatan (Rata-rata Spent($)):")
print(top_earning_categories)

# Visualisasi: 10 Kategori Pekerjaan Teratas Berdasarkan Rata-rata Pelamar
plt.figure(figsize=(12, 7))
sns.barplot(data=top_demand_categories, x='avg_applicants', y='category_1', palette='viridis')
plt.title('10 Kategori Pekerjaan Teratas Berdasarkan Rata-rata Pelamar')
plt.xlabel('Rata-rata Jumlah Pelamar')
plt.ylabel('Kategori Pekerjaan')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Visualisasi: 10 Kategori Pekerjaan Teratas Berdasarkan Rata-rata Spent($)
plt.figure(figsize=(12, 7))
sns.barplot(data=top_earning_categories, x='avg_spent', y='category_1', palette='magma')
plt.title('10 Kategori Pekerjaan Teratas Berdasarkan Rata-rata Pengeluaran Klien')
plt.xlabel('Rata-rata Pengeluaran Klien ($)')
plt.ylabel('Kategori Pekerjaan')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
